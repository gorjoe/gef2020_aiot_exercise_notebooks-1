### Using the RGB Leds

LEDs are output for a system, now let's look at the input. On the PYNQ-Z2 board, there are 2 switches and 4 buttons. Now we will learn how to read the state from the switches.

```python
color = 1 # Try changing this!

for switch_num in range(board.switches.length):
    led_num = switch_num + 4
    
    if board.switches[switch_num].read():
        board.rgbleds[led_num].on(color)
        
    else:
        board.rgbleds[led_num].off()
```
